```python
import matplotlib.pyplot as plt
import math

f1 = open('timesInsertInMap.txt', 'r')
data1 = []
for line in f1:
    data1.append([int(x) for x in line.split()])
f1.close()

f2 = open('timesInsertInAVLTree.txt', 'r')
data2 = []
for line in f2:
    data2.append([int(x) for x in line.split()])
f2.close()

f3 = open('timesInsertInSplayTree.txt', 'r')
data3 = []
for line in f3:
    data3.append([int(x) for x in line.split()])
f3.close()


_, ax = plt.subplots()

xMap = []
yMap = []
xAVL = []
yAVL = []
xSplay = []
ySplay = []
for elem in data1:
    xMap.append(elem[0])
    yMap.append(elem[1])
for elem in data2:
    xAVL.append(elem[0])
    yAVL.append(elem[1])
for elem in data3:
    xSplay.append(elem[0])
    ySplay.append(elem[1])
ax.plot(xMap, yMap, lw = 0.5, color = "r", alpha = 1.0)
ax.plot(xAVL, yAVL, lw = 0.5, color = "b", alpha = 1.0)
ax.plot(xSplay, ySplay, lw = 0.5, color = "g", alpha = 1.0)
pass
```
